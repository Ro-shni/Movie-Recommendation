```python
def recommend(title, sig=sig, indices=indices, movies_cleaned_df=movies_cleaned_df):
    try:
        idx = indices[title]
        sig_scores = list(enumerate(sig[idx]))
        sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
        sig_scores = sig_scores[1:11]
        movie_indices = [i[0] for i in sig_scores]
        recommended_titles = movies_cleaned_df['original_title'].iloc[movie_indices]
        recommended_overviews = movies_cleaned_df['overview'].iloc[movie_indices]
        res = []
        for recommended_title, recommended_overview in zip(recommended_titles, recommended_overviews):
            res.append(f"{recommended_title} ---> {recommended_overview}")
        return res
    except KeyError:
        return f"Movie '{title}' not found in the dataset."
```

```python
movies.dropna(inplace=True)
```

```python
import ast
def convert(text):
  l = []
  for i in ast.literal_eval(text):
    l.append(i['name'])
  return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0]['cast']
```

```python
def convert_cast(text):
  l = []
  counter=0
  for i in ast.literal_eval(text):
    if counter<3:
      l.append(i['name'])
    counter+=1
  return l
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies.iloc[0]['crew']
def convert_crew(text):
  l = []
  counter=0
  for i in ast.literal_eval(text):
    if counter<3:
      l.append(i['name'])
    counter+=1
  return l
  movies['crew']=movies['crew'].apply(convert_cast)
```

```python
movies.iloc[0]['crew']
def fetch_director(text):
  l = []
  for i in ast.literal_eval(text):
    if i['job']=='Director':
      l.append(i['name'])
      break
  return l
  movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
def remove_space(word):
  l = []
  for i in word:
    l.append(i.replace(" ",""))
  return l
```

```python
movies['cast']=movies['cast'].apply(remove_space)
movies['crew']=movies['crew'].apply(remove_space)
movies['genres']=movies['genres'].apply(remove_space)
```
